6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в среднем указывают для каждого типа требуемого опыта (уникальное значение из поля `experience`).

    При решении задачи примите во внимание следующее:
    1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
    2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
    3. Если в расчётах участвует `null`, в результате он тоже даст `null` (посмотрите, что возвращает запрос `select 1 + null`). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит `null` на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

    Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до **целого числа**.

```python
# текст запроса
query_6_6 = f'''
SELECT
    experience,
    Round(AVG(COALESCE((salary_from + salary_to) / 2.0, salary_from, salary_to)), 0) AS average_salary
FROM
    vacancies
WHERE
    (LOWER(name) LIKE '%data scientist%'
   OR LOWER(name) LIKE '%data science%'
   OR LOWER(name) LIKE '%исследователь данных%'
   OR (name like '%ML%' and not name like '%HTML%')
   OR LOWER(name) LIKE '%machine learning%'
   OR LOWER(name) LIKE '%машинн%обучен%')
   AND key_skills IS NOT NULL
   AND key_skills != ''
GROUP BY
    experience
ORDER BY
    experience;
'''
```

```python
# результат запроса
df_6_6 = pd.read_sql_query(query_6_6, connection)
df_6_6
```

- **Сравнительный анализ зарплат по регионам**: можно исследовать, как уровень заработной платы для позиций Data Scientist различается в разных регионах. Это поможет соискателям выбирать наиболее выгодные для себя места для работы.



```python
vacancies = '''(
  (v.name like '%ML%' and not v.name like '%HTML%') or
  (lower(v.name) like '%data scientist%') or
  (lower(v.name) like '%data science%') or
  (lower(v.name) like '%исследователь данных%') or
  (lower(v.name) like '%machine learning%') or
  (lower(v.name) like '%машинн%обучен%')
)'''
```

```python
query_7_1 = f'''
SELECT
    v.area_id,
    a.name,
    Round(AVG(COALESCE((v.salary_from + v.salary_to) / 2.0, v.salary_from, v.salary_to)), 0) AS average_salary
FROM
    vacancies v
    join areas a on v.area_id = a.id

WHERE
    {vacancies}
   and (v.salary_from is not null or v.salary_to is not null)
GROUP BY
    v.area_id, a.name
ORDER BY
    average_salary desc;
'''
```

```python
df_7_1 = pd.read_sql_query(query_7_1, connection)
df_7_1
```

Таким образом, наиболее выгодные по зарплате места работы для соискателя позиции  Data Scientist - это регионы Кипр, Армения, Черногория, Сербия, Москва.

```python
query_7_2 = f'''
SELECT
    COALESCE(a.name, 'Total') AS area_name,
    COUNT(v.id) AS vacancy_count
FROM
    vacancies AS v
JOIN
    employers AS e ON e.id = v.employer_id
JOIN
    areas AS a ON v.area_id = a.id
WHERE
    {vacancies} and
    e.name = 'Яндекс' AND
    a.name IN ({city_list_str})
GROUP BY
    ROLLUP(a.name)
ORDER BY
    vacancy_count;
'''
```

```python
df_7_2 = pd.read_sql_query(query_7_2, connection)
df_7_2
```

Как видно в Яндексе очень мало вакансий для Data Scientist.

```python
query_7_2_1 = f'''
SELECT
    e.name,
    COUNT(*) AS vacancy_count
FROM
    employers e
    JOIN vacancies v ON e.id = v.employer_id

Where
   (v.name like '%ML%' and not v.name like '%HTML%') or
  (lower(v.name) like '%data scientist%') or
  (lower(v.name) like '%data science%') or
  (lower(v.name) like '%исследователь данных%') or
  (lower(v.name) like '%machine learning%') or
  (lower(v.name) like '%машинн%обучен%')

GROUP BY
   e.id
ORDER BY
    vacancy_count DESC
'''
```

```python
df_7_2_1 = pd.read_sql_query(query_7_2_1, connection)
df_7_2_1
```

Таким образом, наиболее заинтересованы в специалистах Data Scientist - это СБЕР, Bell Integrator, Банк ВТБ (ПАО), VK.

```python
query_7_3 = f'''
SELECT
    COALESCE(a.name, 'Total') AS area_name,
    COUNT(v.id) AS vacancy_count
FROM
    vacancies AS v
JOIN
    employers AS e ON e.id = v.employer_id
JOIN
    areas AS a ON v.area_id = a.id
WHERE
    {vacancies} and
    a.name IN ({city_list_str})
GROUP BY
    ROLLUP(a.name)
ORDER BY
    vacancy_count;
'''
```

- **Удаленная работа по регионам**: провести анализ числа вакансий с удаленной работой в различных регионах.

```python
query_7_4 = f'''
SELECT
    v.schedule,
    v.employment,
    COUNT(*) AS vacancy_count
FROM
    vacancies v
Where
   {vacancies}
   and v.schedule = 'Удаленная работа'
GROUP BY
    v.schedule,
    v.employment
ORDER BY
    vacancy_count DESC
'''
```

```python
df_7_4 = pd.read_sql_query(query_7_4, connection)
df_7_4
```
